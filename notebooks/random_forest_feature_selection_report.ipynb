```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
df = pd.read_excel('../outputs/random_forest_feature_selection.xlsx', sheet_name='Sheet1')
```

```python
for col in df.columns[1:-1]:
    font = {'size'   : 14}
    matplotlib.rc('font', **font)
    plt.figure(figsize=(10,7))
    sns.lineplot(x='n_features', y=col,data=df, hue='model_name',style='model_name', markers=True,dashes=True)
    plt.axhline(df[col].max(), linestyle='--',color='black')
    plt.title(col)
    plt.show()
```
