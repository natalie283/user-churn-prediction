```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/natalie/Desktop/DS Thesis/user-churn-prediction/outputs/random_forest_feature_selection.csv')
```

```python
for col in df.columns[1:-1]:
    font = {'size'   : 14}
    matplotlib.rc('font', **font)
    plt.figure(figsize=(10,7))
    sns.lineplot(x='n_features', y=col,data=df, hue='model_name',style='model_name', markers=True,dashes=True)
    plt.axhline(df[col].max(), linestyle='--',color='black')
    plt.title(col, fontsize=24)
    plt.show()
```

Compare feature selection by random forest and logistic regression backward selection

```python
backward_score = pd.read_csv('/Users/natalie/Desktop/DS Thesis/user-churn-prediction/outputs/backward_feature_selection.csv')
```

```python
backward_score['n_features'][backward_score['n_features']=='all']=60
```

```python
backward_score['method'] = 'backward'
backward_score
```

```python
rf_score = df[df['n_features'] != '100']
```

```python
rf_score = rf_score[rf_score['n_features'] != 'all']
```

```python
rf_score['method'] = 'random_forest'
rf_score
```

```python
rf_score.columns
```

```python
union_df = pd.concat([rf_score, backward_score], ignore_index=True)
union_df
```

```python
# Filter the dataframe for model_name 'xgboost'
xgboost_df = union_df[union_df['model_name'] == 'xgboost']

# Convert n_features to numeric type
xgboost_df['n_features'] = pd.to_numeric(xgboost_df['n_features'])

# Ensure the method column is of string type
xgboost_df['method'] = xgboost_df['method'].astype(str)

# Plotting the data
plt.figure(figsize=(10, 6))
sns.lineplot(data=xgboost_df, x='n_features', y='f1_score', hue='method', style='method', markers=True, dashes=False)

# Adding titles and labels
plt.title('F1 Score Comparison of XGBoost Model by Method')
plt.xlabel('Number of Features')
plt.ylabel('F1 Score')
plt.grid(True)
plt.show()
```

```python
xgboost_df
```

```python
xgboost_df = union_df[union_df['model_name'] == 'xgboost']

xgboost_df['n_features'] = pd.to_numeric(xgboost_df['n_features'])

xgboost_df['method'] = xgboost_df['method'].astype(str)

# List of metrics to plot
metrics = ['accuracy_score', 'f1_score', 'auc_score', 'specificity_score', 'sensitivity_score']

# Create a figure with 5 subplots (2 rows, 3 columns)
fig, axs = plt.subplots(2, 3, figsize=(18, 12))
axs = axs.flatten()

# Plot each metric
for i, metric in enumerate(metrics):
    sns.lineplot(data=xgboost_df, x='n_features', y=metric, hue='method', style='method', markers=True, dashes=False, ax=axs[i])
    axs[i].set_title(f'{metric.replace("_", " ").title()} ', fontsize=24)
    axs[i].set_xlabel('Number of Features')
    axs[i].set_ylabel(metric.replace('_', ' ').title())
    axs[i].grid(True)

# Hide the empty subplot
fig.delaxes(axs[-1])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# model logistic regression

m_df = union_df[union_df['model_name'] == 'log_reg']
m_df['n_features'] = pd.to_numeric(m_df['n_features'])
m_df['method'] = m_df['method'].astype(str)
metrics = ['accuracy_score', 'f1_score', 'auc_score', 'specificity_score', 'sensitivity_score']

# Create a figure with 5 subplots (2 rows, 3 columns)
fig, axs = plt.subplots(2, 3, figsize=(18, 12))
axs = axs.flatten()

# Plot each metric
for i, metric in enumerate(metrics):
    sns.lineplot(data=m_df, x='n_features', y=metric, hue='method', style='method', markers=True, dashes=False, ax=axs[i])
    axs[i].set_title(f'{metric.replace("_", " ").title()} LR-Comparison')
    axs[i].set_xlabel('Number of Features')
    axs[i].set_ylabel(metric.replace('_', ' ').title())
    axs[i].grid(True)

# Hide the empty subplot
fig.delaxes(axs[-1])

# Adjust layout
plt.tight_layout()
plt.show()
```
